```python
import numpy as np
import pandas as pd
import codecs
import re
```

```python
lines = []
with codecs.open('./data/cornell movie-dialogs corpus/movie_lines.txt', 'r', encoding="utf-8", errors="ignore") as lines_f:
    lines_reader = lines_f.read().split("\n")
    for line in lines_reader:
        lines.append(line.split(' +++$+++ '))
```

```python
movie_lines = pd.DataFrame(lines[:-1], columns=['line_code', 'character_code', 'movie_code', 'character_name', 'line'])
movie_lines['line_code'].replace(regex=True, inplace=True, to_replace=r'\D', value=r'')
movie_lines['line_code'] = pd.to_numeric(movie_lines['line_code'])
movie_lines
```

```python
sequence = movie_lines[['line_code', 'line']].sort_values('line_code')
sequence
```

```python
lines_dict = {}
for line in lines:
    if len(line) > 4:
        key = line[0][1:]
        dialogue = line[4]
        lines_dict[int(key)] = dialogue
```

```python
lines_dict
```

```python
sorted_lines = sorted(lines_dict.items(), key = lambda x: x[0])
sorted_lines
```

```python
sequence = {}
counter = 1
chat_id = []
for i in range(1, len(sorted_lines)+1):
    if i < len(sorted_lines):
        if (sorted_lines[i][0] - sorted_lines[i-1][0]) == 1:
            if sorted_lines[i-1][1] not in chat_id:
                chat_id.append(sorted_lines[i-1][1])
            chat_id.append(sorted_lines[i][1])
        elif (sorted_lines[i][0] - sorted_lines[i-1][0]) > 1:            
            sequence[counter] = chat_id
            chat_id = []
        counter += 1
    else:
        pass
```

```python
sequence
```

```python
context_and_target = []
for chat in sequence.values():
    if len(chat) % 2 != 0:
        chat = chat[:-1]
    for i in range(0, len(chat), 2):
        context_and_target.append((chat[i], chat[i+1]))
```

```python
context_and_target
```

```python
context, target = zip(*context_and_target)
context = list(context)
target = list(target)
```

```python
context
```

```python
target
```

```python
def clean_text(text):
    '''Clean text by removing unnecessary characters and altering the format of words.'''

    text = text.lower()
    
    text = re.sub(r"i'm", "i am", text)
    text = re.sub(r"he's", "he is", text)
    text = re.sub(r"she's", "she is", text)
    text = re.sub(r"it's", "it is", text)
    text = re.sub(r"that's", "that is", text)
    text = re.sub(r"what's", "that is", text)
    text = re.sub(r"where's", "where is", text)
    text = re.sub(r"how's", "how is", text)
    text = re.sub(r"\'ll", " will", text)
    text = re.sub(r"\'ve", " have", text)
    text = re.sub(r"\'re", " are", text)
    text = re.sub(r"\'d", " would", text)
    text = re.sub(r"\'re", " are", text)
    text = re.sub(r"won't", "will not", text)
    text = re.sub(r"can't", "cannot", text)
    text = re.sub(r"n't", " not", text)
    text = re.sub(r"n'", "ng", text)
    text = re.sub(r"'bout", "about", text)
    text = re.sub(r"'til", "until", text)
    text = re.sub(r"[-()\"#/@;:<>{}`+=~|.!?,]", "", text)
    
    return text
```

```python
tidy_target = []
for conve in target:
    text = clean_text(conve)
    tidy_target.append(text)
tidy_context = []
for conve in context:
    text = clean_text(conve)
    tidy_context.append(text)
```

```python
bos = "<BOS> "
eos = " <EOS>"
final_target = [bos + conve + eos for conve in tidy_target] 
encoder_inputs = tidy_context
decoder_inputs = final_target
```
